# Boombike Case Study
### Problem Statement

A US bike sharing provider Boombikes aspires to understand the demand for shared bikes among the people.
They want to prepare themselves to cater to the people's needs once the situation gets better all around after corona pandemic 
and stand out from other service providers and make huge profits.

The company wants to know:
- Which variables are significant in predicting the demand for shared bikes.
- How well those variables describe the bike demands

### Model builiding
1) Split the dataset in two: train and test dataframes
2) Scale the training dataset using fit_transform- min max scaling
3) Define cnt as dependent(y) variable
4) Define other columns as independent(X) variables
5) Using linear regression RFE select top 15 features
6) Train the model using least square method
7) Check the coefficients and intercepts
8) Consider alpha/Significant value = 0.05
9) Check P-value, Rsquared value
10) Check VIF
11) Take manual approach by eliminating column with high p-value and high VIF
12) Take steps from number 6 to 11 again.

```python
df_train, df_test = train_test_split(bike_sharing, train_size =0.7, random_state = 100 )
print(df_train.shape)
print(df_test.shape)
```

```python
scaler = MinMaxScaler()
#create list of numeric variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
#fit on data
#scaler.fit--- learns/ computes xmax and xmin
#scaler.transform----x-xmin/xmax-xmin
#scaler.fit_transform---fit and transform 

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
X_train.head()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
dtypes = {col: 'float64' for col in X_train_rfe.columns}
X_train_rfe = X_train_rfe.astype(dtypes)
```

```python
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values,i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
X_train_rfe.drop('season_spring', axis = 1, inplace = True)
```

```python
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values,i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
X_train_rfe.drop('mnth_Oct', axis = 1, inplace = True)
```

```python
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values,i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
lm.params
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(X_train_rfe.corr(),annot = True, cmap= "YlGnBu")
plt.show()
```

### Residual analysis on train dataset

```python
y_train_pred = lm.predict(X_train_rfe)
```

```python
y_train_pred
```

```python
res = y_train - y_train_pred
sns.distplot(res)
```

### Prediction on the test dataset

```python
num_vars = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
y_test = df_test.pop('cnt')
```

```python
X_test = df_test
```

```python
X_test_rfe = X_test[['yr', 'holiday', 'temp', 'hum', 'windspeed', 'mnth_Jan', 'mnth_Jul',
       'mnth_Oct', 'mnth_Sep', 'season_spring', 'season_summer',
       'season_winter', 'weekday_Tue', 'weathersit_Light_Rain',
       'weathersit_Mist']]
X_test_rfe = sm.add_constant(X_test_rfe)
```

```python
X_test_rfe.head()
```

```python
dtypes = {col: 'float64' for col in X_test_rfe.columns}
X_test_rfe = X_test_rfe.astype(dtypes)
```

```python
X_test_rfe.drop(['season_spring','mnth_Oct'],axis = 1, inplace = True)
```

```python
y_test_pred = lm.predict(X_test_rfe)
```

### Model evaluation

```python
r2_score(y_true = y_test, y_pred = y_test_pred)
```

```python
fig = plt.figure()
sns.regplot(x = y_test, y = y_test_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_test_pred', fontsize=16)                     # Y-label
plt.show()
```

```python
import pylab as py
res = y_test - y_test_pred
res_np = np.array(res)
plt.figure(figsize = (16,5))
sns.distplot(res)
plt.title('residual normal distribution')
sm.qqplot(res_np,line = 's')
plt.title('Q-Q plot')
```
